```python
import chromadb
from dotenv import load_dotenv
load_dotenv()
import os
api_key=os.getenv('OPENAI_API_KEY')
```

```python
client = chromadb.PersistentClient(path="./chroma_db")
```

```python
collection = client.get_or_create_collection(name="collection")
```

https://platform.openai.com/docs/guides/embeddings

```python
import openai
def get_openai_embedding(text):
    response=openai.embeddings.create(model= "text-embedding-3-small",
    input= [text],)
    return response.data[0].embedding
```

```python
documents = ["eiffel tower in paris",
"taj mahal is famous",
"python is programming language",
"colosseum is in rome"]
```

```python
embeddings = [get_openai_embedding(doc) for doc in documents]
```

```python
len(embeddings[1])
```

Adding embedding documents to chromadb collections

```python
collection.add(
    ids=[str(i) for i in range(len(documents))],
    documents=documents,
    embeddings=embeddings
)
```

```python
query_text = "where is Eiffel tower?"
query_embedding = get_openai_embedding(query_text)
results=collection.query(
    query_embeddings=[query_embedding],
    n_results=2,
    include=["documents","distances"]
)

print(results["documents"][0])
print(results["distances"][0])
```
